```python
import os
from google.colab import drive

# 1. Montar Drive
drive.mount('/content/drive')

def list_files(startpath, prefix=''):
    """Genera una estructura de Ã¡rbol visual para una ruta dada."""
    if not os.path.exists(startpath):
        print(f"La ruta no existe: {startpath}")
        return

    files = sorted(os.listdir(startpath))
    for i, file in enumerate(files):
        path = os.path.join(startpath, file)
        is_last = (i == len(files) - 1)

        connector = 'â””â”€â”€ ' if is_last else 'â”œâ”€â”€ '
        print(f"{prefix}{connector}{file}")

        if os.path.isdir(path):
            extension = '    ' if is_last else 'â”‚   '
            list_files(path, prefix + extension)

# 2. Configurar la ruta
# Para carpetas compartidas, primero debes aÃ±adir un "Acceso directo" (Shortcut)
# a tu 'My Drive' para que sea visible desde la ruta de Colab.
ruta_carpeta = '/content/drive/MyDrive/PJLA 2025'

print(f"ğŸ“‚ {os.path.basename(ruta_carpeta)}")
list_files(ruta_carpeta)
```
