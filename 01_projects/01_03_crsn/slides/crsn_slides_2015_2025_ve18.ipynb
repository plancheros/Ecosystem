```python
# ==============================================================================
# BLOQUE 1: ETL & AUDITOR√çA DE DATOS (STAGING AREA)
# PROYECTO: CRSN MASTER 2015-2025 | Versi√≥n: vPJLA_v5.7_AUDIT
# ==============================================================================

import pandas as pd
import numpy as np
import os

def ejecutar_bloque_1_hardcore(excel_path, output_folder):
    years = range(2015, 2026)
    stats = {"success": [], "warnings": [], "errors": []}
    diezmos_list = []
    siervos_list = []

    # Lista Maestra de Ministerios (Est√°ndar 18)
    mins_targets = [
        "UJIERES", "ALABANZA", "VISA", "NICODEMO", "ROCA KIDS", "ORACION",
        "MUJERES", "HOMBRES", "CASADOS", "TMT", "J25", "A√ëOS DORADOS",
        "EJECUTIVOS", "COMUNICACIONES", "ROCAFE", "CLUBES", "JOSUES", "CLUB DE AMIGOS"
    ]

    print(f"üïµÔ∏è  AUDITOR√çA INICIADA: Procesando matriz 2015-2025...")
    print("-" * 50)

    for year in years:
        try:
            # Carga de hoja
            df = pd.read_excel(excel_path, sheet_name=str(year), header=None)

            # --- ETL DIEZMOS ---
            d_idx = df[df.apply(lambda r: r.astype(str).str.fullmatch('DIEZMOS', case=False).any(), axis=1)].index
            f_diezmos = False
            if not d_idx.empty:
                vals = pd.to_numeric(df.iloc[d_idx[0], 1:13], errors='coerce').fillna(0).tolist()
                diezmos_list.append([year] + vals + [sum(vals)])
                f_diezmos = True
            else:
                diezmos_list.append([year] + [0]*13)

            # --- ETL SIERVOS ---
            s_header = df[df.apply(lambda r: r.astype(str).str.contains('SIERVOS', case=False).any(), axis=1)].index
            y_data = {"A√ëO": year}
            f_siervos = False

            if not s_header.empty:
                start = s_header[0] + 1
                for m_name in mins_targets:
                    m_row = df.iloc[start:start+30, 0].astype(str).str.contains(m_name.replace('+', '\\+'), case=False, na=False)
                    if m_row.any():
                        r_idx = m_row[m_row].index[0]
                        m_vals = pd.to_numeric(df.iloc[r_idx, 1:13], errors='coerce').fillna(0)
                        y_data[m_name] = m_vals.max() # Tomamos el pico para resiliencia
                        f_siervos = True
                    else:
                        y_data[m_name] = 0
            siervos_list.append(y_data)

            # Registro de progreso
            if f_diezmos and f_siervos:
                stats["success"].append(year)
                print(f"   ‚úÖ {year}: Datos completos.")
            else:
                stats["warnings"].append(year)
                print(f"   ‚ö†Ô∏è  {year}: Datos parciales encontrados.")

        except Exception as e:
            stats["errors"].append(f"{year}: {str(e)}")
            print(f"   ‚ùå {year}: ERROR CR√çTICO -> {e}")

    # --- PERSISTENCIA (Generaci√≥n de CSVs) ---
    df_d = pd.DataFrame(diezmos_list, columns=['A√ëO'] + [f'MES_{i}' for i in range(1,13)] + ['TOTAL'])
    # Calcular √çndice 100
    base_2015 = df_d.loc[df_d['A√ëO']==2015, 'TOTAL'].values[0]
    df_d['INDICE'] = (df_d['TOTAL'] / (base_2015 if base_2015 > 0 else 1)) * 100

    df_s = pd.DataFrame(siervos_list)

    # Guardar en la carpeta incremental
    path_d = os.path.join(output_folder, "stg_diezmos.csv")
    path_s = os.path.join(output_folder, "stg_siervos.csv")
    df_d.to_csv(path_d, index=False)
    df_s.to_csv(path_s, index=False)

    # ==========================================================================
    # PRINTOUT DE RESULTADOS FINALES
    # ==========================================================================
    print("\n" + "‚ïê"*50)
    print("üìä REPORTE FINAL DE INGENIER√çA - BLOQUE #1")
    print("‚ïê"*50)
    print(f"‚úÖ √âXITO TOTAL:  {len(stats['success'])} a√±os")
    print(f"‚ö†Ô∏è  ADVERTENCIAS: {len(stats['warnings'])} a√±os")
    print(f"‚ùå ERRORES:      {len(stats['errors'])} a√±os")
    print("-" * 50)

    if len(stats['errors']) == 0:
        print("üöÄ ESTADO GLOBAL: SUCCESSFUL - CAPA DE STAGING LISTA")
    else:
        print("üö® ESTADO GLOBAL: FAILED - REVISAR LOGS DE ERROR")

    print(f"üìÇ ARCHIVOS: {os.path.basename(path_d)} | {os.path.basename(path_s)}")
    print("‚ïê"*50 + "\n")

    return df_d, df_s

# LLAMADA AL MOTOR
df_diezmos, df_siervos = ejecutar_bloque_1_hardcore(EXCEL_PATH, OUTPUT_FOLDER)
```

```python
# -----------------------------------------------------------------------------
# Bloque: #2 - Matriz de Diezmos (Gradiente Azul Inverso & Borde Oro)
# Est√°ndar: PJLA Hardcore Architecture (Starlink Visual Grade)
# -----------------------------------------------------------------------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import os

def ejecutar_heatmap_diezmos_v26_29():
    print("üé® [BLOQUE 2] Renderizando Mapa de Calor: GRADIENTE ADAPTATIVO...")

    # 1. Recuperaci√≥n de Data (Staging)
    path_stg = os.path.join(OUTPUT_FOLDER, "stg_diezmos.csv")
    df_raw = pd.read_csv(path_stg)
    meses_cols = [f'MES_{i}' for i in range(1, 13)]

    # Normalizaci√≥n √çndice 100 (Base 2015) para confidencialidad
    base_2015 = df_raw.loc[df_raw['A√ëO'] == 2015, meses_cols].values.mean()
    df_plot = (df_raw.set_index('A√ëO')[meses_cols].copy() / base_2015) * 100
    df_plot.columns = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN',
                       'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']

    # 2. Configuraci√≥n del Canvas (Maximizando el 95% del √°rea)
    fig = plt.figure(figsize=(16, 9), facecolor="#000028")
    ax = fig.add_axes([0.08, 0.1, 0.88, 0.68])
    ax.set_facecolor("#000028")

    # 3. Renderizado con Colores Complementarios (Azul/Oro)
    sns.heatmap(df_plot, cmap="Blues", cbar=False,
                linewidths=2.5, linecolor="#FFD700", ax=ax)

    # 4. L√≥gica Antag√≥nica: Contraste Adaptativo de Datos (95% Espacio)
    for y in range(df_plot.shape[0]):
        for x in range(df_plot.shape[1]):
            val = int(df_plot.iloc[y, x])
            # Si el fondo es oscuro (val alto), texto blanco. Si es claro, texto navy.
            t_color = "white" if val > 220 else "#000028"
            outline_color = "black" if t_color == "white" else "white"

            txt = ax.text(x + 0.5, y + 0.5, f"{val}",
                          ha='center', va='center', color=t_color,
                          fontsize=26, fontweight='black')

            # Filete (Outline) de 3pt para legibilidad Starlink
            txt.set_path_effects([path_effects.withStroke(linewidth=3, foreground=outline_color)])

    # 5. T√≠tulos y Branding (Ajuste a Proporci√≥n √Åurea)
    plt.figtext(0.53, 0.90, "CRSN: DIEZMOS 2015-2025", color="#FFD700",
                fontsize=54, ha='center', weight='black')
    plt.figtext(0.53, 0.83, "2015: Base 100", color="#FFFFFF",
                fontsize=32, ha='center', style='italic')

    # Guardado de Activo
    img_path = os.path.join(OUTPUT_FOLDER, "02_diezmos_heatmap_final.png")
    plt.savefig(img_path, dpi=300, bbox_inches='tight', facecolor="#000028")
    plt.show()

ejecutar_heatmap_diezmos_v26_29()
```

```python
# -----------------------------------------------------------------------------
# Autor:       Pedro J. Lancheros / Peter Boat
# Proyecto:    CRSN Slides (3J / VitaNova)
# Bloque:      #3.2 - Master Heatmap Engine (Blueprint v26.22)
# Fecha:       2026-01-30
# Hora:        21:05 (-5 UTC Bogot√°)
# Versi√≥n:     v26.22_20260130_2105 (Starlink Grade)
# -----------------------------------------------------------------------------

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from datetime import datetime

# --- 1. ESPECIFICACIONES T√âCNICAS (BLUEPRINT v26.22) ---
COLOR_NAVY = "#000028"      # 60% Fondo
COLOR_ORO = "#FFD700"       # 10% T√≠tulos (Oro Vital)
COLOR_BLANCO = "#FFFFFF"    # Datos / Subt√≠tulos
PHI = (1 + 5**0.5) / 2      # Proporci√≥n √Åurea

# --- 2. PREPARACI√ìN DE MATRIZ (STAGING DATA) ---
# Usamos los datos auditados inyectados anteriormente
meses_labels = ["ENE", "FEB", "MAR", "ABR", "MAY", "JUN", "JUL", "AGO", "SEP", "OCT", "NOV", "DIC"]

# Normalizaci√≥n Intra-a√±o: Cada fila (a√±o) se divide por su propio promedio y x100
df_m_norm = df_audit_heatmap.copy() # Asumiendo que viene del Bloque 2 o Inyecci√≥n

# --- 3. RENDERIZADO HEATMAP (EST√ÅNDAR CRSN) ---
plt.figure(figsize=(16, 9))
plt.rcParams.update({'figure.facecolor': COLOR_NAVY, 'axes.facecolor': COLOR_NAVY})

# Paleta: Escala de Ocre a Naranja Quemado (YlOrBr)
ax = sns.heatmap(df_m_norm,
                 annot=True,
                 fmt="d",
                 cmap="YlOrBr",
                 cbar=False,
                 linewidths=1.5,
                 linecolor=COLOR_NAVY,
                 annot_kws={"weight": "black", "size": 14}) # L√≥gica Antag√≥nica: N√∫meros Negros para contraste

# Est√©tica de Ejes
plt.title("DIEZMOS: MAPA DE CALOR (ESTACIONALIDAD)\nPromedio Mensual del A√±o = 100",
          color=COLOR_ORO, fontsize=32, pad=30, fontweight='bold')
plt.xticks(color=COLOR_BLANCO, fontsize=16, fontweight='bold')
plt.yticks(color=COLOR_BLANCO, fontsize=16, fontweight='bold', rotation=0)

# Guardar PNG de alta fidelidad
path_heat_v26 = os.path.join(OUTPUT_FOLDER, "heatmap_v26_22.png")
plt.savefig(path_heat_v26, dpi=300, bbox_inches='tight', facecolor=COLOR_NAVY)
plt.show()

# --- 4. EXPORTACI√ìN PPTX (ANCLAJE √ÅUREO) ---
prs = Presentation()
prs.slide_width, prs.slide_height = Inches(13.33), Inches(7.5)

def add_vn_heatmap_slide(title, sub, img):
    slide = prs.slides.add_slide(prs.slide_layouts[6])
    slide.background.fill.solid()
    slide.background.fill.fore_color.rgb = RGBColor(0, 0, 40) # Navy #000028

    # Logo (Anclaje x=0.01)
    if os.path.exists(LOGO_PATH):
        slide.shapes.add_picture(LOGO_PATH, Inches(0.1), Inches(0.1), height=Inches(1.2))

    # T√≠tulo (54pt, Oro Vital, $H/\phi^2$)
    t_box = slide.shapes.add_textbox(Inches(0), Inches(7.5/(PHI**2)), Inches(13.33), Inches(1))
    t = t_box.text_frame.paragraphs[0]
    t.text = title; t.alignment = 1
    t.font.bold, t.font.size = True, Pt(54)
    t.font.color.rgb = RGBColor(255, 215, 0)

    # Subt√≠tulo (32pt, Blanco, It√°lica)
    s_box = slide.shapes.add_textbox(Inches(0), Inches(7.5/(PHI**2) + 0.35), Inches(13.33), Inches(0.8))
    s = s_box.text_frame.paragraphs[0]
    s.text = sub; s.alignment = 1
    s.font.italic, s.font.size = True, Pt(32)
    s.font.color.rgb = RGBColor(255, 255, 255)

    # Imagen del Heatmap (Ocupaci√≥n M√°xima)
    slide.shapes.add_picture(img, Inches(0.4), Inches(2.5), width=Inches(12.5), height=Inches(4.8))

# Ejecuci√≥n de la l√°mina
add_vn_heatmap_slide("ESTACIONALIDAD MENSUAL", "An√°lisis de Intensidad - CRSN", path_heat_v26)

final_pptx_heat = os.path.join(OUTPUT_FOLDER, f"AUDITORIA_HEATMAP_v26_22.pptx")
prs.save(final_pptx_heat)

print("-" * 40)
print(f"‚úÖ Bloque #3.2 (v26.22): EXITOSO")
print(f"üìÑ Heatmap Slide: {os.path.basename(final_pptx_heat)}")
print("-" * 40)
```
