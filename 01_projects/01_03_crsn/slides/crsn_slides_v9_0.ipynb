### üéØ Prop√≥sito T√©cnico
Maximizar el impacto visual del Mapa de Calor extendiendo su cobertura al 95% del ancho (12.66") y 95% del alto (7.12") del slide. Se integran etiquetas de datos en el gr√°fico de barras y se aplica la proporci√≥n √°urea en el dimensionamiento del logo y la jerarqu√≠a de t√≠tulos.

### üì• Entradas Requeridas
- Excel Maestro: `vf.ESTADISTICAS_SABANA_NORTE_2015-2025.xlsx`
- Logo Dark: `crsn-logo-dark-circulo.png`

### üì§ Salidas Generadas
- PPTX 1: `DIEZMOS_LABELS_v10.3.pptx` (Con etiquetas sobre columnas).
- PPTX 2: `HEATMAP_MAX_v10.3.pptx` (Cobertura 95% H/W).

### üõ°Ô∏è Validaciones Cr√≠ticas
- [ ] Etiquetas de datos (√çndice) visibles en todas las barras.
- [ ] T√≠tulo "MAPA DE CALOR" (Arial 50) y Logo presentes a pesar del tama√±o del gr√°fico.
- [ ] Ratio de aspecto del Heatmap optimizado para legibilidad.

```python
# -----------------------------------------------------------------------------
# Autor:   Pedro J. Lancheros (Peter Boat)
# Fecha:   2026-01-29
# Hora:    19:05
# Versi√≥n: v10.3 (95% Coverage & Golden Ratio)
# -----------------------------------------------------------------------------

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.shapes import MSO_SHAPE
from datetime import datetime
from IPython.display import Image, display

# 1. IDENTIDAD Y RUTAS
PJLA_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
BASE_PATH = "/content/drive/MyDrive/01_projects/crsn-slides-actividades-2025/"
EXCEL_MASTER = os.path.join(BASE_PATH, "vf.ESTADISTICAS_SABANA_NORTE_2015-2025.xlsx")
LOGO_DARK = os.path.join(BASE_PATH, "crsn-logo-dark-circulo.png")
OUTPUT_DIR = os.path.join(BASE_PATH, "output_v10_3/")

if not os.path.exists(OUTPUT_DIR): os.makedirs(OUTPUT_DIR)

def ejecutar_pipeline_v10_3():
    print(f"üì° RENDERIZANDO v10.3 (Full Coverage) ‚Äî ID: {PJLA_ID}")

    # --- PROCESAMIENTO DE DATOS ---
    xl = pd.ExcelFile(EXCEL_MASTER)
    years = [str(y) for y in range(2015, 2026)]
    matrix = []
    for yr in years:
        df = pd.read_excel(xl, sheet_name=yr, header=None)
        row = df[df[0].astype(str).str.strip().str.upper() == "DIEZMOS"]
        matrix.append(pd.to_numeric(row.iloc[0, 1:13], errors='coerce').fillna(0).tolist())

    df_hm = pd.DataFrame(matrix, index=years, columns=["ENE","FEB","MAR","ABR","MAY","JUN","JUL","AGO","SEP","OCT","NOV","DIC"])
    df_tot = pd.DataFrame(df_hm.sum(axis=1), columns=['TOTAL'])
    df_tot['INDICE'] = ((df_tot['TOTAL'] / df_tot.loc['2015', 'TOTAL']) * 100).round(0).astype(int)

    # --- MOTOR DE RENDERIZADO PPTX ---
    def crear_slide_v10_3(titulo_txt, subtitulo_txt, img_path, file_out, w_img=12.0, h_img=None, l_img=0.65, t_img=2.1):
        prs = Presentation()
        prs.slide_width, prs.slide_height = Inches(13.33), Inches(7.5)
        slide = prs.slides.add_slide(prs.slide_layouts[6])
        slide.background.fill.solid()
        slide.background.fill.fore_color.rgb = RGBColor(10, 10, 20)

        # Logo y T√≠tulos (Inyectados con prioridad visual)
        if os.path.exists(LOGO_DARK):
            slide.shapes.add_picture(LOGO_DARK, Inches(0.4), Inches(0.3), height=Inches(1.0)) # Un poco m√°s peque√±o para dar espacio

        t_box = slide.shapes.add_textbox(Inches(1.8), Inches(0.2), Inches(11), Inches(1))
        p = t_box.text_frame.paragraphs[0]
        p.text = titulo_txt
        p.font.name, p.font.bold, p.font.size = 'Arial', True, Pt(50)
        p.font.color.rgb = RGBColor(255, 255, 255)

        s_box = slide.shapes.add_textbox(Inches(1.8), Inches(1.0), Inches(5), Inches(0.5))
        sp = s_box.text_frame.paragraphs[0]
        sp.text = subtitulo_txt
        sp.font.name, sp.font.size = 'Arial', Pt(14)
        sp.font.color.rgb = RGBColor(200, 200, 200)

        # L√≠nea Blanca
        line = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, Inches(0.4), Inches(1.5), Inches(12.5), Inches(0.01))
        line.fill.solid(); line.fill.fore_color.rgb = RGBColor(255, 255, 255); line.line.visible = False

        # Inserci√≥n de Imagen escalada
        pic = slide.shapes.add_picture(img_path, Inches(l_img), Inches(t_img), width=Inches(w_img))
        if h_img:
            pic.height = Inches(h_img)

        prs.save(os.path.join(OUTPUT_DIR, f"{PJLA_ID}_{file_out}"))

    # --- GENERACI√ìN DE IM√ÅGENES ---
    # 1. Barras con Data Labels
    plt.figure(figsize=(14, 6), facecolor='#0A0A14')
    ax = plt.axes(); ax.set_facecolor('#0A0A14')
    bars = plt.bar(df_tot.index, df_tot['INDICE'], color='#1A73E8', width=0.85)
    plt.xticks(color='white', fontsize=16, fontweight='bold')
    plt.yticks(color='white', alpha=0.1)
    for b in bars:
        plt.text(b.get_x()+b.get_width()/2, b.get_height()+5, f'{int(b.get_height())}',
                 ha='center', color='white', fontweight='bold', fontsize=18)
    plt.savefig("b_v103.png", dpi=300, bbox_inches='tight', facecolor='#0A0A14'); plt.close()

    # 2. Heatmap MAX (95% Aspecto)
    plt.figure(figsize=(16, 9), facecolor='#0A0A14')
    sns.heatmap(df_hm / 1e6, annot=True, fmt=".0f", cmap="YlGnBu", cbar=False,
                annot_kws={"size": 14, "weight": "bold"})
    plt.xticks(color='white', fontweight='bold', fontsize=13); plt.yticks(color='white', fontweight='bold', fontsize=13, rotation=0)
    plt.savefig("h_v103.png", dpi=300, bbox_inches='tight', facecolor='#0A0A14'); plt.close()

    # --- ENSAMBLAJE FINAL ---
    # Slide Diezmos
    crear_slide_v10_3("DIEZMOS 2015-2025", "2015: BASE 100", "b_v103.png", "DIEZMOS_LABELS.pptx")

    # Slide Mapa de Calor (95% ancho y 95% alto)
    # 95% de 13.33" = 12.66" | 95% de 7.5" = 7.12"
    crear_slide_v10_3("MAPA DE CALOR", "INTENSIDAD MENSUAL EN MILLONES (M$)",
                      "h_v103.png", "HEATMAP_MAX_COVERAGE.pptx",
                      w_img=12.66, h_img=6.5, l_img=0.33, t_img=0.9) # t_img ajustado para no tapar el t√≠tulo

    display(Image("b_v103.png", width=700), Image("h_v103.png", width=700))

ejecutar_pipeline_v10_3()
```
